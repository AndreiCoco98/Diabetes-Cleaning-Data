# Diabetes Data Cleaning Project

 The dataset should be cleaned in the following steps:
 
 1. Checking the number of columns
 2. Changing misspelt column names
 3. Checking for missing values
 4. Checking fror duplicate values
 5. Checking for possible outliers

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os as os
```

```python
# Read file
df = pd.read_csv(r"C:\Users\coco MANEL\Documents\Pandas\diabetes_unclean.csv")
df
```

```python
# Quick look at the head records of the table
df.head()
```

```python
# Checking the number of columns
df.columns
```

```python
# Rename column name "No_Pation" to 'Patients_No'
```

```python
df.rename(columns= {'No_Pation':'Patients_No'}, inplace = True)
```

```python
# Check if column changed
df
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Replace missing value in column "HbA1c"

mean_value = df['HbA1c'].mean()
mean_value
```

```python
# Filling missing values with the mean

df['HbA1c'] = df['HbA1c'].fillna(mean_value).round(2)
```

```python
# Confirm that the change was done correctly
df.isnull().sum()
```

```python
# Dropping the missing values from other columns
df1 = df.dropna(how = 'any')
df1.isnull().sum()
```

```python
# Check the information concerning the dataset

df.info()
```

```python
df1.groupby('CLASS')['CLASS'].agg("count")
```

```python
# Check unique values in "CLASS" column

df['CLASS'].unique()
```

```python
# Remove blanks from the row in order to solve the inconsistency

df['CLASS'] = df['CLASS'].str.strip()
```

```python
df['CLASS'].unique()
```

```python
# Check for outliers in dataset

sns.boxplot(df1["Cr"])
```

```python
# Choose maximum quanttile to fill in the ooutliers
max_cr = df['Cr'].quantile(0.995)
max_cr
```

```python
df1[df1['Cr'] > max_cr] 
```

```python
# Assign new data to the outliers to a new dataframe

df2 = df1[df1['Cr'] < max_cr]
```

```python
# Confirm the change

sns.boxplot(df2['Cr'])
```

```python
# Checking for outlier in column "HbA1c"

sns.boxplot(df2['HbA1c'], orient = 'h')
```

```python
# Check for duplicate values

df2.duplicated()
```

```python
# Drop duplicated values

df3 = df2.drop_duplicates()
```

```python
# Confirm the change

df3
```

```python
# Saving the dataframe into a new csv file

df3.to_csv("Cleaned_diabetes_data.csv")
```
